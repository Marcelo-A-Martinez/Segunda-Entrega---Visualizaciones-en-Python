```python
# GRÁFICO DE TORTA
# En este caso se determinó la cantidad de pacientes de acuerdo al género,
# luego se procedio a indicar dichos valores y crear la gráfica respectiva.
```

```python
pacientes_genero = Clinica['genero'].value_counts()
print(pacientes_genero)
```

```python
x = [57643 , 42357]
labels = ['FEMENINO' , 'MASCULINO']
colors = ['tab:orange',  'tab:blue']

fig, ax = plt.subplots()
ax.pie(x, labels = labels, 
       colors = colors, autopct='%.0f%%')
ax.set_title('Porcentaje de pacientes según el género')
plt.show()
```

```python
# Interpretación:
# Como se puede observar en el gráfico de torta "Porcentaje de pacientes según el género", el 58% de los pacientes
# corresponden al género femenino y el restante 42% al masculino; esto nos indica que en mayor proporción
# los pacientes son mujeres.
```

```python
# GRÁFICO DE BARRAS
# En este caso, en primer lugar se creo la variable "CANTIDAD_PACIENTES_ESTADIA"
# donde se contabilizó la cantidad de pacientes en función de los días de estancia en la clínica.
# Posteriormente, estos valores se volcaron en la gráfica correspondiente.
```

```python
CANTIDAD_PACIENTES_ESTADIA = Clinica.groupby(['duracionEstancia']).size().reset_index(name='Cantidad_Pacientes')
print (CANTIDAD_PACIENTES_ESTADIA)
```

```python
fig, ax = plt.subplots(figsize=(8,4))
plt.xticks(range(18))
ax.bar(CANTIDAD_PACIENTES_ESTADIA['duracionEstancia'], CANTIDAD_PACIENTES_ESTADIA.Cantidad_Pacientes)
ax.set_title('Cantidad de pacientes según la duración de la estadía')
ax.set_ylabel('Cantidad de Pacientes')
ax.set_xlabel('Duración de la Estadía (en días)')
```

```python
# interpretación:
# Observando el gráfico de barras se puede establecer que los pacientes permanecen en la clínica
# entre 1 y 6 días. Además, se puede indicar que en general una cantidad importante de pacientes permanece
# solamente un día (1).
```

```python
# BOXPLOT
# Para realizar este gráfico, se utilizaron los datos contenidos en la variable "CANTIDAD_PACIENTES_ESTADIA"
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.boxplot(x=CANTIDAD_PACIENTES_ESTADIA.T,)
ax.set_title('Boxplot de la cantidad de pacientes según la duración de la estadía')
ax.set_xlabel('Duración de la Estadía (en días)')
ax.set_ylabel('Cantidad de Pacientes')
```

```python
#Interpretación:
# Este grafico boxplot complementa al anterior, ya que se puede observar los valores mínimos y máximos
# de la cantidad de pacientes en funcón de la duración de su estadía en la clínica.
# También permite apreciar la media, la mediana y los cuartiles.
# Además, no se visualizan valores atípicos en el mismo.
```

```python
# GRÁFICO DE DISPERSIÓN
# Se procedió a relacionar los atributos de bmi (indice de masa corporal) y la respiración.
```

```python
ax = sns.scatterplot(data=Clinica, x="bmi", y="respiracion" )
ax.set (xlabel = 'Índice de masa corporal (bmi)',
       ylabel = 'Frecuencia respiratoria',
       title = 'BMI de los pacientes vs. Frecuencia respiratoria')
```

```python
#Interpretación:
# En el gráfico de dispersión no se puede establecer una tendencia en cuanto a la relación entre las variables
# índice de masa corporal de los pacientes y su correspondiente frecuencia respiratoria.
```

```python
# HISTOGRAMA
# El mismo se diseñó teniendo en cuenta el atributo "respiracion" del dataset "Clinica".
# Se contabilizó la cantidad de pacientes en base a los registros de respiración de los mismos.
```

```python
ax = sns.histplot(data=Clinica, x="respiracion", bins=5)
ax.set (xlabel = 'Frecuencia Respiratoria',
       ylabel = 'Cantidad de Pacientes',
       title = 'Respiración de los pacientes')
```

```python
#Interpretación:
# En este caso se puede ver que la mayoría de los pacientes presentan una frecuencia respiratoria 
# entre 6 y 8.
```

```python
# BOXPLOT
# En este caso se generó el boxplot a partir del género de los pacientes (Femenino y Masculino)
# y la cantidad de días que permanecieron en la clínica.
```

```python
ax = sns.boxplot(x="genero", y="duracionEstancia", data=Clinica, palette="coolwarm")
plt.legend(labels=["0 = Masculino","1 = Femenino"], 
                   title = "Género" , 
                   loc = 2, 
                   bbox_to_anchor = (1,1) , 
                   fontsize='12')
ax.set (xlabel = 'Género',
       ylabel = 'Duración de la estadía (en días)',
       title = 'Boxplot de la duración de la estadía según el género de los pacientes')
```

```python
# Interpretación:
# En este boxplot se puede apreciar el tiempo de estadía de los pacientes según el género de los mismos.
# Se puede establecer que los hombres suelen permanecer más tiempo en la clínica en contraposición a
# las mujeres.
# Además, en ambos casos se observan valores atípicos, en especial en el caso de las mujeres.
```
